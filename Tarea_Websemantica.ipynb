Celda de Transformación (CSV a RDF)

```python
from rdflib import Graph, Literal, RDF, URIRef, Namespace
from rdflib.namespace import FOAF, SDO

# Definición de espacios de nombres (Namespaces)
CLINIC = Namespace("http://www.clinica-bienestar.es/ontology#")
g = Graph()
g.bind("sdo", SDO)
g.bind("clinic", CLINIC)

# Cargar el CSV (Asegúrate de que el archivo esté en la ruta data/raw/)
df = pd.read_csv('data/raw/hospitales.csv')

for index, row in df.iterrows():
    hospital_uri = URIRef(f"http://www.clinica-bienestar.es/data/hospital/{row['CNH']}")

    g.add((hospital_uri, RDF.type, SDO.Hospital))
    g.add((hospital_uri, SDO.name, Literal(row['HOSPITAL'])))
    g.add((hospital_uri, SDO.identifier, Literal(row['CNH'])))

    # Agregar dirección
    address_uri = URIRef(f"http://www.clinica-bienestar.es/data/address/{row['CNH']}")
    g.add((hospital_uri, SDO.address, address_uri))
    g.add((address_uri, RDF.type, SDO.PostalAddress))
    g.add((address_uri, SDO.streetAddress, Literal(row['DIRECCION'])))
    g.add((address_uri, SDO.postalCode, Literal(row['COD_POSTAL'])))
    g.add((address_uri, SDO.addressLocality, Literal(row['MUNICIPIO'])))

# Guardar el resultado en la carpeta de procesados
g.serialize(destination='data/processed/hospitales.ttl', format='turtle')
print("✅ Transformación a Turtle completada.")
```
